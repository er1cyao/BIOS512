```r
library(dplyr)

step1_initialize_clusters <- function(df, k) {
  df$cluster <- sample(paste0("C", 0:(k - 1)), nrow(df), replace = TRUE)
  return(df)
}

step2_compute_centroids <- function(df) {
  numeric_cols <- df %>% select(where(is.numeric), -cluster) %>% colnames()
  df %>%
    group_by(cluster) %>%
    summarise(across(all_of(numeric_cols), mean, .names = "{.col}"), .groups = "drop") %>%
    arrange(cluster) %>%
    mutate(label = paste0("μ", row_number() - 1))
}

step3_assign_clusters <- function(df, centroids) {
  numeric_cols <- df %>% select(where(is.numeric), -cluster) %>% colnames()
  for (i in 1:nrow(df)) {
    distances <- sapply(1:nrow(centroids), function(j) {
      sqrt(sum((df[i, numeric_cols] - centroids[j, numeric_cols])^2))
    })
    df$cluster[i] <- centroids$cluster[which.min(distances)]
  }
  return(df)
}

step4_check_convergence <- function(old_clusters, new_clusters) {
  all(old_clusters == new_clusters)
}

my_kmeans <- function(df, k, max_iter = 100) {
  df <- step1_initialize_clusters(df, k)
  iter <- 1
  converged <- FALSE
  
  while (iter <= max_iter && !converged) {
    cat("Iteration", iter, "...\n")
    old_clusters <- df$cluster
    centroids <- step2_compute_centroids(df)
    df <- step3_assign_clusters(df, centroids)
    converged <- step4_check_convergence(old_clusters, df$cluster)
    iter <- iter + 1
  }
  
  return(list(labels = df$cluster, means = centroids))
}
```
